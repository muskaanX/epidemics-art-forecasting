necessary libraries

```python
import pandas as pd
from pathlib import Path
```

loading files for drugs administered with RTV booster

```python
atv = pd.read_csv('datasets/atv_adult_pre_calibration.csv')
drv = pd.read_csv('datasets/drv_adult_pre_calibration.csv')
lpv = pd.read_csv('datasets/lpv_pediatric_pre_calibration.csv')
```

summing patient counts across boosted regimens

```python
rtv_booster = atv[["month", "year"]].copy()
rtv_booster["patients"] = atv["patients"] + drv["patients"] + lpv["patients"]
```

saving

```python
DATASETS = Path("datasets")
rtv_booster.to_csv(DATASETS / "rtv_booster_pre_calibration.csv", index=False)
```
