necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

loading the interpolated data

```python
INTERPOLATED = Path("/Users/muskaanchugh/Desktop/art_synthetic_data_gen/interpolated_data")
interpolated_dfs = {f.stem.replace("_synthetic_interpolated", ""): pd.read_csv(f) for f in INTERPOLATED.glob("*.csv")}
```

defining analysis function

```python
def analyze_drug(name, df):
    df = df.copy()
    df["date"] = pd.to_datetime(df["month"] + " " + df["year"].astype(str))
    df = df.sort_values("date")
    df["mom_change"] = df["patients"].diff()

    print(f"mean:      {df['patients'].mean():.0f}")
    print(f"median:    {df['patients'].median():.0f}")
    print(f"std:       {df['patients'].std():.0f}")
    print(f"min:       {df['patients'].min()}")
    print(f"max:       {df['patients'].max()}")
    print(f"range:     {df['patients'].max() - df['patients'].min()}")
    print(f"cv:        {df['patients'].std() / df['patients'].mean():.4f}")
    print(f"mom mean:  {df['mom_change'].mean():.1f}")
    print(f"mom std:   {df['mom_change'].std():.1f}")
    peak = df.loc[df['patients'].idxmax()]
    trough = df.loc[df['patients'].idxmin()]
    print(f"peak:      {peak['patients']} ({peak['date'].strftime('%b %Y')})")
    print(f"trough:    {trough['patients']} ({trough['date'].strftime('%b %Y')})")

    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 6))
    ax1.plot(df["date"], df["patients"])
    ax1.set_title(f"{name} — patient count")
    ax1.set_ylabel("patients")
    ax2.plot(df["date"], df["mom_change"])
    ax2.axhline(0, color="red", linestyle="--", linewidth=0.8)
    ax2.set_title(f"{name} — month-on-month change")
    ax2.set_ylabel("change")
    plt.tight_layout()
    plt.show()
```

abc+3tc - adult

```python
analyze_drug("abc_3tc — adult", interpolated_dfs["abc_3tc_adult"])
```

```python
analyze_drug("atv/r — adult", interpolated_dfs["atv_adult"])
```
